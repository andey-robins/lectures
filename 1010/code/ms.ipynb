```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = 0.05

k = 1.0
m = 2.0

phi = 0.0
```

```python
print(A)
```

```python
omega = np.sqrt(k/m)

t = np.arange(0,10,0.1)
```

```python
x = A * np.cos(omega * t + phi)
```

```python
print(x)
```

```python
plt.plot(t,x)
```

```python
volume = 10
moles = 2
R = 8.314 # constant 
temp = 283
print((moles * R * temp) / volume)
```

```python
y_pressure = []
t = np.arange(273, 273+100, 1) 
for temperature in t:
    temp = temperature
    y_pressure.append((moles * R * temp) / volume)
```

```python
plt.plot(t,y_pressure)
```
